````python
import pandas as pd

# Load Excel (simulating internal collection)
file_path = "/content/Marketing_Campaign_Advanced_Internship.xlsx"

customers = pd.read_excel(file_path, sheet_name="Customer_Master")
campaigns = pd.read_excel(file_path, sheet_name="Campaign_Interactions")
outcomes = pd.read_excel(file_path, sheet_name="Campaign_Outcome")

# Save structured CSVs
customers.to_csv("customer_master.csv", index=False)
campaigns.to_csv("campaign_interactions.csv", index=False)
outcomes.to_csv("campaign_outcome.csv", index=False)

print("Data collected and saved.")
````

````python
# Load CSVs
customers = pd.read_csv("customer_master.csv")
campaigns = pd.read_csv("campaign_interactions.csv")
outcomes = pd.read_csv("campaign_outcome.csv")

# Remove duplicates
customers.drop_duplicates(inplace=True)
campaigns.drop_duplicates(inplace=True)
outcomes.drop_duplicates(inplace=True)

# Handle invalid values
customers = customers[customers["Annual_Income_INR"] > 0]
campaigns = campaigns[campaigns["Campaign_Cost_INR"] > 0]

# Feature engineering
campaigns["CTR"] = campaigns["Clicks"] / campaigns["Impressions"]
# Cap CTR values to be between 0 and 1
campaigns["CTR"] = campaigns["CTR"].clip(0, 1)

# Merge datasets for cleaning consistency
df = campaigns.merge(customers, on="Customer_ID") \
              .merge(outcomes, on="Campaign_ID")

# Data validation checks
assert customers["Customer_ID"].is_unique
assert campaigns["Campaign_ID"].is_unique
assert outcomes["Campaign_ID"].is_unique
assert df["CTR"].between(0, 1).all()

# Save cleaned data
df.to_csv("cleaned_marketing_data.csv", index=False)
print(" Data cleaned and validated.")
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv("cleaned_marketing_data.csv")

# Descriptive stats
print(df.describe())

# ROI by Campaign Type
roi_summary = df.groupby("Campaign_Type")["ROI_INR"].mean()
print("Average ROI by Campaign Type:\n", roi_summary)

# Conversion rate by Campaign Type
conversion_rate = df.groupby("Campaign_Type")["Purchase_Made"].mean()
print("Conversion Rate by Campaign Type:\n", conversion_rate)

# Discount effect on ROI and conversion
discount_analysis = df.groupby("Discount_Applied_Percent")[["Purchase_Made", "ROI_INR"]].mean()
print("Discount impact:\n", discount_analysis)
````

````python
# Example: identify top 3 campaigns by ROI
top_campaigns = df.groupby("Campaign_ID")["ROI_INR"].sum().sort_values(ascending=False).head(3)
print("Top 3 campaigns by total ROI:", top_campaigns)
````

````python
# Conversion by Loyalty Tier and Device
segment_analysis = df.groupby(["Loyalty_Tier","Device_Type"])["Purchase_Made"].mean()
print(segment_analysis)
````

````python
#KPI'S

avg_roi = df['ROI_INR'].mean()
avg_conversion = df['Purchase_Made'].mean()
avg_ctr = df['CTR'].mean()
total_revenue = df['Net_Revenue_INR'].sum()

print(f"\nAverage ROI: {avg_roi:.2f}")
print(f"Average Conversion Rate: {avg_conversion*100:.2f}%")
print(f"Average CTR: {avg_ctr:.2f}")
print(f"Total Net Revenue: INR {total_revenue:,.0f}")
````

````python
df = df.sort_values("Campaign_ID").reset_index(drop=True)
df['Campaign_Date'] = pd.date_range(start='2023-01-01', periods=len(df), freq='15D')
df['Month'] = df['Campaign_Date'].dt.strftime('%b')  # Month as Jan, Feb, etc.
df['Month_Num'] = df['Campaign_Date'].dt.month  # For sorting


#  Monthly KPI Trends

monthly_kpis = df.groupby(['Month', 'Month_Num']).agg({
    'ROI_INR':'mean',
    'Purchase_Made':'mean',
    'CTR':'mean',
    'Net_Revenue_INR':'sum'
}).rename(columns={'Purchase_Made':'Conversion_Rate'}).sort_values('Month_Num')

# Plot trends
plt.figure(figsize=(10,5))
plt.plot(monthly_kpis.index.get_level_values('Month'), monthly_kpis['ROI_INR'], marker='o')
plt.title("Monthly Average ROI Trend")
plt.xlabel("Month")
plt.ylabel("ROI")
plt.grid(True)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(monthly_kpis.index.get_level_values('Month'), monthly_kpis['Conversion_Rate'], marker='o', color='green')
plt.title("Monthly Conversion Rate Trend")
plt.xlabel("Month")
plt.ylabel("Conversion Rate")
plt.grid(True)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(monthly_kpis.index.get_level_values('Month'), monthly_kpis['CTR'], marker='o', color='orange')
plt.title("Monthly CTR Trend")
plt.xlabel("Month")
plt.ylabel("CTR")
plt.grid(True)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(monthly_kpis.index.get_level_values('Month'), monthly_kpis['Net_Revenue_INR'], marker='o', color='purple')
plt.title("Monthly Net Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Net Revenue (INR)")
plt.grid(True)
plt.show()


# 3Ô∏è Top Campaign Type Trend Over Time
-
# Select top 3 campaign types by overall ROI
top_types = df.groupby("Campaign_Type")["ROI_INR"].mean().sort_values(ascending=False).head(3).index
campaign_monthly_top = df[df["Campaign_Type"].isin(top_types)]
campaign_monthly_top = campaign_monthly_top.groupby(['Month','Month_Num','Campaign_Type']).agg({
    'ROI_INR':'mean',
    'Purchase_Made':'mean'
}).rename(columns={'Purchase_Made':'Conversion_Rate'}).reset_index().sort_values('Month_Num')

plt.figure(figsize=(10,5))
sns.lineplot(data=campaign_monthly_top, x='Month', y='ROI_INR', hue='Campaign_Type', marker='o')
plt.title("Top 3 Campaign Type ROI Trend Over Time")
plt.xlabel("Month")
plt.ylabel("ROI")
plt.grid(True)
plt.show()

plt.figure(figsize=(10,5))
sns.lineplot(data=campaign_monthly_top, x='Month', y='Conversion_Rate', hue='Campaign_Type', marker='o')
plt.title("Top 3 Campaign Type Conversion Rate Trend Over Time")
plt.xlabel("Month")
plt.ylabel("Conversion Rate")
plt.grid(True)
plt.show()


#  Insights

top_months_roi = monthly_kpis.reset_index()[['Month','ROI_INR']].sort_values('ROI_INR', ascending=False).head(3)
top_months_conversion = monthly_kpis.reset_index()[['Month','Conversion_Rate']].sort_values('Conversion_Rate', ascending=False).head(3)

print("\nTop 3 months by ROI:\n", top_months_roi)
print("\nTop 3 months by Conversion Rate:\n", top_months_conversion)
````

````python
# Campaign Type Summary
campaign_summary = df.groupby("Campaign_Type").agg({
    "ROI_INR":"mean",
    "Purchase_Made":"mean",
    "CTR":"mean"
}).rename(columns={"Purchase_Made":"Conversion_Rate"}).sort_values("ROI_INR", ascending=False)
print("\nCampaign Type Summary:\n", campaign_summary)

# Top 5 campaigns by ROI
top_campaigns = df.groupby("Campaign_ID")["ROI_INR"].sum().sort_values(ascending=False).head(5)
top_campaigns_df = top_campaigns.reset_index()
print("\nTop 5 Campaigns by ROI:\n", top_campaigns_df)

# Campaign Objective Impact
objective_summary = df.groupby("Campaign_Objective").agg({
    "ROI_INR":"mean",
    "Purchase_Made":"mean"
}).rename(columns={"Purchase_Made":"Conversion_Rate"})
print("\nCampaign Objective Impact:\n", objective_summary)

# Conversion Rate by Loyalty Tier
loyalty_summary = df.groupby("Loyalty_Tier", observed=True)["Purchase_Made"].mean().reset_index()
loyalty_summary.rename(columns={"Purchase_Made":"Conversion_Rate"}, inplace=True)
print("\nConversion Rate by Loyalty Tier:\n", loyalty_summary)

# Average Purchase Value by Age
age_purchase = df.groupby("Age", observed=True)["Purchase_Value_INR"].mean().reset_index()
print("\nAverage Purchase Value by Age:\n", age_purchase.head(10))

# Gender-wise Performance
gender_summary = df.groupby("Gender", observed=True).agg({
    "Purchase_Made":"mean",
    "ROI_INR":"mean"
}).rename(columns={"Purchase_Made":"Conversion_Rate"}).reset_index()
print("\nGender-wise Performance:\n", gender_summary)

# Income Segment Analysis
income_bins = [0, 300000, 600000, 1000000, 2000000]
income_labels = ["<3L", "3-6L", "6-10L", ">10L"]
df["Income_Segment"] = pd.cut(df["Annual_Income_INR"], bins=income_bins, labels=income_labels)
income_summary = df.groupby("Income_Segment", observed=True).agg({
    "ROI_INR":"mean",
    "Purchase_Made":"mean"
}).rename(columns={"Purchase_Made":"Conversion_Rate"}).reset_index()
print("\nIncome Segment Analysis:\n", income_summary)

# Conversion Rate by Device Type
device_summary = df.groupby("Device_Type", observed=True)["Purchase_Made"].mean().reset_index()
device_summary.rename(columns={"Purchase_Made":"Conversion_Rate"}, inplace=True)
print("\nConversion Rate by Device Type:\n", device_summary)

# Average ROI by Time of Day
time_summary = df.groupby("Time_of_Day", observed=True)["ROI_INR"].mean().reset_index()
print("\nAverage ROI by Time of Day:\n", time_summary)
````
